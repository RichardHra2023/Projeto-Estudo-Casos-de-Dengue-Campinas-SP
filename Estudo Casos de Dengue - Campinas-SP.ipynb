```python
# Verificando a quantidade de dados nulos por coluna
df.isnull().sum()
```

```python
# Total de casos confirmados no estudo
df['casos-confirmados'].sum()
```

```python
# Extraindo o ano e o mês
df['Mes'] = pd.DatetimeIndex(df['data']).month
df['Ano'] = pd.DatetimeIndex(df['data']).year
```

```python
df
```

```python
# Somando a quantidade de casos confirmados por ano
df_agrupado_ano = df.groupby('Ano').sum('casos-confirmados')
```

```python
df_agrupado_ano
```

```python
df_agrupado_ano.reset_index()
```

### Podemos ver no gráfico abaixo que houve um aumento muito grande de casos de dengue no ano de 2014

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 8))

x = df_agrupado_ano.index  # Se 'Ano' é o índice, você pode acessá-lo diretamente
y = df_agrupado_ano['casos-confirmados']

plt.plot(x, y)
plt.title('Casos Confirmados ao Longo dos Anos (Agrupados)')
plt.xlabel('Ano', fontsize = 14)
plt.ylabel('Casos Confirmados', fontsize = 14)
plt.grid(True)
plt.show()
```
